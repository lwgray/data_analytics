# Color

As we saw last week, matplotlib offers us a lot of different ways to customize the colors in our plots. 
- Pass the color name or its abbreviation as a string, such as `"blue"` or `"b"`.
- Hex color codes can also be specified using its HTML/CSS hex string, such as `"#0000FF"` for the color blue.
- Instead of a color, a gray level can be passed as a string for a float value in `[0, 1]`, with `"0.0"` being black and `"1.0"` white. 

## Colormaps

Matplotlib supports a wide range of colormaps that allow us to translate numerical data values into visible colors in a plot.

Remember from last week's class, the type of color palette we want to use depends on our data. The three main types of color palettes are qualitative, sequential, and diverging.

**Matplotlib Reference Guides**  
Matplotlib Colormaps [Tutorial](https://matplotlib.org/stable/tutorials/colors/colormaps.html)  
Choosing Colormaps in Matplotlib [Guide](https://matplotlib.org/stable/tutorials/colors/colormaps.html?highlight=colormaps)  
Matplotlib Colormaps Reference [documentation](https://matplotlib.org/stable/gallery/color/colormap_reference.html?highlight=colormaps%20reference)  

```python
# Get the current color cycle
cycle_colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
cycle_colors
```

```python
cmaps = [('Perceptually Uniform Sequential', [
            'viridis', 'plasma', 'inferno', 'magma', 'cividis']),
         ('Sequential', [
            'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']),
         ('Sequential (2)', [
            'binary', 'gist_yarg', 'gist_gray', 'gray', 'bone', 'pink',
            'spring', 'summer', 'autumn', 'winter', 'cool', 'Wistia',
            'hot', 'afmhot', 'gist_heat', 'copper']),
         ('Diverging', [
            'PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu',
            'RdYlBu', 'RdYlGn', 'Spectral', 'coolwarm', 'bwr', 'seismic']),
         ('Cyclic', ['twilight', 'twilight_shifted', 'hsv']),
         ('Qualitative', [
            'Pastel1', 'Pastel2', 'Paired', 'Accent',
            'Dark2', 'Set1', 'Set2', 'Set3',
            'tab10', 'tab20', 'tab20b', 'tab20c']),
         ('Miscellaneous', [
            'flag', 'prism', 'ocean', 'gist_earth', 'terrain', 'gist_stern',
            'gnuplot', 'gnuplot2', 'CMRmap', 'cubehelix', 'brg',
            'gist_rainbow', 'rainbow', 'jet', 'nipy_spectral', 'gist_ncar'])]

gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

def plot_color_gradients(cmap_category, cmap_list):
    # Instantiate Figure and adjust height to number of colormaps
    nrows = len(cmap_list)
    figh = 0.35 + 0.15 + (nrows + (nrows-1)*0.1)*0.22
    fig, axes = plt.subplots(nrows=nrows, figsize=(6.4, figh))
    fig.subplots_adjust(top=1-.35/figh, bottom=.15/figh, left=0.2, right=0.99)

    axes[0].set_title(cmap_category + ' colormaps', fontsize=14)
    for ax, name in zip(axes, cmap_list):
        ax.imshow(gradient, aspect='auto', cmap=plt.get_cmap(name))
        ax.text(-.01, .5, name, va='center', ha='right', fontsize=10, transform=ax.transAxes)

    # Turn off ticks and spines
    for ax in axes:
        ax.set_axis_off()

for cmap_category, cmap_list in cmaps:
    plot_color_gradients(cmap_category, cmap_list)
```

In this example, we want to use one of matplotlib's diverging colormaps. We can set the colormap by passing its name as a string to the `cmap` parameter.

# Text

We can easily add text to our plots by using both matplotlib's object-oriented interface and the pyplot module. For both methods, we first pass the location of the x and y coordinates, and then the text as a string. 

> `ax.text(x, y, "text string")`
> `plt.text(x, y, "text string")`

In addition, take a look at the following example to see all of the different ways we can customize our text! 

**Matplotlib Methods**  
`matplotlib.axes.Axes.text` [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.text.html)  
`matplotlib.pyplot.text` [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html)

```python
# Plot the data
ax = tips['total_bill'].plot()

# Add text (left-aligned) using OOP interface
ax.text(40, 50, 'This text starts at point (40,50)', color='tab:purple', fontsize=12)

# Add right-aligned text with the pyplot module
plt.text(100, 5, 'This text ends at point (100,5)',
         style='italic',
         ha='right');
#          horizontalalignment='right');

# Add a bounding box around the text with the pyplot module
plt.text(20, 40, 'Bounding box', bbox={'facecolor': 'grey', 'alpha': 0.5, 'pad': 10});
```

# Major & Minor Ticks

For each plot or Axes object, both the x-axis and y-axis have major and minor tick marks. However, by default, matplotlib will hide the minor ticks, except when drawing logarithmic plots. In addition, 

Matplotlib by default will also add tick labels to the plot's left and bottom axis. We can change this by using the `tick_params()` method to turn the ticks and the tick labels on and off for each axis independently. 

The parameters for the major ticks are `bottom`, `left`, `top`, and `right`. By default, `left` and `bottom` are set to `True`. For tick labels, the arguments are `labelbottom`, `labelleft`, `labeltop`, and `labelright`. You can have one without the other, both, or neither on any axis in any combination.

**Matplotlib Methods & Reference Guides**  
Matplotlib Major and Minor Tick Marks [Documentation](https://matplotlib.org/stable/gallery/ticks_and_spines/major_minor_demo.html)  
`matplotlib.axes.Axes.get_xticklabels` [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.get_xticklabels.html)  
`matplotlib.axes.Axes.get_yticklabels` [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.get_yticklabels.html)  
`matplotlib.axes.Axes.tick_params` [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.tick_params.html)

```python
# Instantiate the Figure and a 1x3 grid
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(14,7))

# Plot with the defaults
tips['total_bill'].plot(ax=ax1)

# Remove bottom ticks and add ticks to top and right
tips['total_bill'].plot(ax=ax2)
ax2.tick_params(bottom=False, top=True, right=True)

# Remove the bottom and left labels and add top and right labels
tips['total_bill'].plot(ax=ax3)
ax3.tick_params(labelbottom=False, labelleft=False, labeltop=True, labelright=True);
```

Now, let's see how we can format our tick marks using the `tick_params()` method!

**Matplotlib Methods**  
`matplotlib.axes.Axes.tick_params` [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.tick_params.html)

```python
# Plot the bar chart
ax = tips['total_bill'].plot(figsize=(6,3))

# Customize the major xticks
ax.tick_params(axis='x', direction='out', length=16, width=4, color='xkcd:teal')
ax.tick_params(axis='y', direction='in', length=6, width=8, color='xkcd:coral')
```

# Grids

Adding gridlines to your plot can help the viewer to discern specific data values and to compare values in one part of the chart to values in another part. However, you should try to use gridlines sparingly since they can add visual noise and even cause the moire effect that we saw the first day in class when they are densely spaced.

**Matplotlib Methods**  
`matplotlib.axes.Axes.grid` [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.grid.html)

```python
# Create some example data
x = np.linspace(-6, 6, 500)
y = np.sinc(x)

# Instantiate the Figure and Axes object
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6))

# Plot the data
ax1.plot(x,y)

# Add grid to the x- and y-axis
ax1.grid(axis='both')

# Plot the data
ax2.plot(x,y)

# Add and customize grids individually
ax2.grid(axis='x', color='xkcd:maroon', alpha=.6, linewidth=2, linestyle=':')
ax2.grid(axis='y', color='black', alpha=.5, linewidth=.5);
```

By default, matplotlib places the grid on top of our other chart elements. However, we can change this by setting `set_axisbelow(True)`.

**Matplotlib Methods**  
`matplotlib.axes.Axes.set_axisbelow` [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_axisbelow.html)

```python
# Instantiate the Figure and Axes object
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(14,7))

# Plot the data
titanic['age'].plot.hist(ax=ax1, edgecolor='xkcd:black')

# Add grid to the x- and y-axis
ax1.grid(axis='y')

# Plot the data
titanic['age'].plot.hist(ax=ax2, edgecolor='xkcd:black')

# Add grid to the x- and y-axis
ax2.grid(axis='y')

# Don't allow the axis to be on top of your data
ax2.set_axisbelow(True);
```

# Arrows

Annotation is another powerful way to customize our plots! We can use it to call out surprising features and add additional information that can help turn out plot into a story.

The pattern for calling out an annotation has several parts, and they're all mixed up from what we've seen before. We first pass the text that we want to include to the `annotate()` method. Next, a tuple containing the x and y coordinates of where we want our arrow to end. The next tuple should contain the coordinates of where we want our text to start.

Finally, we can specify that we want to plot an arrow by using the `arrowprops` argument. It takes a dictionary of the possible arrow properties. By setting `arrowstyle="-|>"`, matplotlib will plot a triangular arrow head at the non-text end.

**Matplotlib Methods**  
`matplotlib.axes.Axes.annotate` [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html)

```python
# Instantiate the Figure and Axes object
fig, ax = plt.subplots(figsize=(9,6))

# Plot the data
ax.plot(x,y)

# Add a plain arrow
ax.annotate('Annotation text', (0, 1), (-4, 0.5), arrowprops={'arrowstyle': '-|>'})

# Add a fancy arrow
ax.annotate('Fancy arrow', (0, 1), (2, 0.5), arrowprops={'arrowstyle': 'fancy'});
```
