```python
# insert code
import pandas as pd
mtc = pd.read_csv('monster_trading_cards.csv')
mtc.head()
```

# Top Spenders
## Identify the top 5 spenders in the game by total purchase value, and provide the following info on each player.
* user_name
* Purchase count
* Average Purchase Price
* Total Purchase Value

```python
mtc.groupby('user_name')['price'].agg(['count', 'mean', 'sum']).sort_values(by='sum', ascending=False).head()
```

# Most Popular Items

## Identify the top 5 most popular `monster card` by purchase count, then list (in a table):
  * monster card
  * Purchase Count
  * Average Item Price
  * Total Purchase Value
  * The max price paid for this card
  * The min price paid for this card

```python
mtc.groupby('monster_card')['price'].agg(['count', 'mean', 'sum', 'min', 'max']).sort_values('count', ascending=False).head()
```

# Most Profitable Items

## Identify the top 5 most profitable monster_card by total purchase value, then list (in a table):
  * Monster Card
  * Purchase Count
  * Average Item Price
  * Total Purchase Value

```python
mtc.groupby('monster_card')['price'].agg(['count', 'mean','sum']).sort_values('sum', ascending=False).head()
```

# DateTime
Determine how many passengers flew in the Third Quarter of 2017?
* Use the `AirPassengers.csv` dataset
* Subset only dates in 1950 into a new DataFrame
* Group new DataFrame into Quarters and aggregate the `#Passengers` column with aggregating function `sum` then sort Passengers in descending order

```python
ap = pd.read_csv('AirPassengers.csv', parse_dates=['Month'], index_col='Month')
```

```python
year_1950 = ap.loc['1950']
```

```python
year_1950.groupby(pd.Grouper(freq='Q')).sum()
```
